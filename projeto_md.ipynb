# 1. Business Understanding

## 1.1 Descrição do problema de negócio

O crescimento do uso de redes sociais, plataformas digitais e ferramentas baseadas em inteligência artificial tem transformado significativamente a rotina dos estudantes, influenciando aspectos acadêmicos, sociais e psicológicos. Embora essas tecnologias ofereçam benefícios relacionados à comunicação, aprendizado e acesso à informação, o uso excessivo pode estar associado a problemas como redução da qualidade do sono, dificuldades de concentração, aumento de ansiedade e impactos no desempenho acadêmico.

Nesse contexto, compreender como diferentes padrões de uso tecnológico afetam o cotidiano estudantil tornou-se uma questão relevante para instituições de ensino, pesquisadores e profissionais da área educacional e de saúde digital. Entretanto, devido à grande quantidade de variáveis comportamentais presentes nos dados e à complexidade das relações entre elas, identificar padrões manualmente é uma tarefa difícil.

Assim, o projeto propõe a utilização de técnicas de Mineração de Dados para explorar o dataset, identificar perfis de estudantes com comportamentos digitais semelhantes e investigar relações entre hábitos tecnológicos, saúde mental, qualidade do sono e desempenho acadêmico. Além das técnicas de clustering para descoberta de grupos, também serão utilizados modelos de regressão linear múltipla para analisar quantitativamente a influência de diferentes variáveis sobre indicadores de bem-estar e produtividade acadêmica.

Os resultados esperados incluem a geração de insights sobre comportamento digital entre estudantes, a identificação de possíveis fatores de risco associados ao uso excessivo de tecnologia e a produção de análises interpretáveis que possam contribuir para futuras pesquisas e estratégias de conscientização sobre bem-estar digital.

- **Contexto:** Dataset composto por informações relacionadas ao uso de redes sociais, hábitos digitais, desempenho acadêmico, qualidade do sono e indicadores de saúde mental de estudantes. O conjunto de dados busca representar diferentes perfis de comportamento tecnológico em contexto educacional.
-  **Problema:** Dificuldade em compreender como diferentes padrões de uso de tecnologia e redes sociais se relacionam com desempenho acadêmico, bem-estar psicológico e qualidade do sono, especialmente devido à presença de múltiplas variáveis comportamentais e subjetivas.
- **Possíveis interessados:** Pesquisadores das áreas de educação, saúde digital e ciência de dados, instituições de ensino, psicólogos educacionais, gestores acadêmicos e estudantes interessados em compreender os impactos do uso de tecnologia no cotidiano acadêmico.
- **Decisões apoiadas pela análise:** 
desenvolvimento de estratégias de conscientização sobre uso saudável de tecnologia;
identificação de perfis de estudantes mais vulneráveis a impactos negativos do uso excessivo de redes sociais;
geração de insights para futuras pesquisas sobre comportamento digital, saúde mental e desempenho acadêmico;
compreensão de fatores associados à qualidade do sono e produtividade acadêmica.
- **Limitações iniciais conhecidas:** 
O dataset é baseado em respostas autorreportadas, podendo conter vieses subjetivos ou inconsistências nas informações fornecidas pelos participantes.
Algumas variáveis categóricas podem apresentar distribuição desbalanceada entre categorias.
A existência de correlações não implica causalidade entre uso de tecnologia e impactos acadêmicos ou psicológicos.
Técnicas de clustering podem gerar grupos sensíveis à escolha de parâmetros, normalização dos dados e seleção de atributos.
Modelos de regressão linear podem ser afetados por multicolinearidade, outliers e relações não lineares entre variáveis.

## 1.2 Objetivo do dataset, origem e características gerais

### Objetivo do dataset

O dataset tem como objetivo analisar e enetender o impacto do uso de inteligência artificial, redes sociais e plataformas digitais na vida acadêmica e pessoal de estudantes. Os dados incluem informações sobre comportamento digital, características demográficas, desempenho acadêmico, saúde mental, qualidade do sono e percepção dos impactos da tecnologia, permitindo investigar padrões de uso e possíveis relações entre hábitos tecnológicos, sono e bem-estar. 

### Origem dos dados

O conjunto de dados foi disponibilizado publicamente na plataforma Kaggle, sendo voltado para análises relacionadas ao impacto das tecnologias digitais e redes sociais na saúde e rotina estudantil.

Link público: https://www.kaggle.com/datasets/sumeakash/impact-of-social-media-on-health  

### Características do dataset

| Coluna | Tipo esperado | Descrição | Relevância para a tarefa |
|---|---|---|---|
| Student_ID | Categórica / Identificador | Identificador único do estudante | Utilizada apenas para identificação, sem relevância analítica direta |
| Age | Numérica | Idade do estudante (em anos) | Auxilia na análise de padrões comportamentais por faixa etária |
| Gender | Categórica | Gênero do estudante (Male/Female) | Permite avaliar diferenças de comportamento entre grupos |
| Academic_Level | Categórica | Nível acadêmico do estudante (School / College / University)| Importante para comparar hábitos digitais em diferentes etapas acadêmicas |
| Country | Categórica | País de origem do estudante | Possibilita análises exploratórias entre diferentes contextos geográficos |
| Avg_Daily_Usage_Hours | Numérica | Média diária de uso de plataformas digitais | Variável central para análise da intensidade de uso tecnológico |
| Most_Used_Platform | Categórica | Plataforma digital mais utilizada | Permite identificar preferências e perfis de comportamento |
| Affects_Academic_Performance | Booleana | Impacto percebido da tecnologia no desempenho acadêmico | Relacionada diretamente aos objetivos da investigação |
| Sleep_Hours_Per_Night | Numérica | Média de horas de sono por noite | Indicador importante de qualidade de vida e bem-estar |
| Mental_Health_Score | Numérica | Indicador relacionado à saúde mental do estudante | Essencial para investigar possíveis impactos psicológicos |
| Relationship_Status | Categórica | Situação de relacionamentos pessoais | Pode auxiliar na análise de aspectos sociais relacionados ao uso digital |
| Conflicts_Over_Social_Media | Categórica | Existência de conflitos relacionados ao uso de redes sociais | Indicador de possíveis impactos sociais do uso excessivo |
| Overall_Impact | Categórica | Avaliação geral do impacto da tecnologia na vida do estudante (Negative/Positive/Neutral) | Importante para interpretação dos perfis encontrados |
### Relação com o problema de negócio

O dataset é adequado para o problema proposto porque reúne informações diretamente relacionadas aos hábitos digitais, desempenho acadêmico, saúde mental e qualidade de vida dos estudantes. A diversidade de variáveis numéricas e categóricas permite aplicar técnicas de clustering para identificar perfis comportamentais e modelos de regressão linear múltipla para investigar relações quantitativas entre uso de tecnologia e indicadores de bem-estar. Além disso, o conjunto de dados possibilita análises interpretáveis e exploratórias alinhadas ao objetivo de compreender os impactos do uso intenso de plataformas digitais no contexto educacional.
