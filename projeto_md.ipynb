```python
# Visualizações exploratórias básicas
# O grupo deve adaptar as colunas às características do dataset.

def plot_distribuicao_numerica(df: pd.DataFrame, coluna: str) -> None:
    plt.figure(figsize=(8, 4))
    df[coluna].dropna().hist(bins=30)
    plt.title(f"Distribuição de {coluna}")
    plt.xlabel(coluna)
    plt.ylabel("Frequência")
    plt.tight_layout()
    plt.show()

def plot_contagem_categorica(df: pd.DataFrame, coluna: str, top_n: int = 20) -> None:
    plt.figure(figsize=(10, 4))
    df[coluna].value_counts(dropna=False).head(top_n).plot(kind="bar")
    plt.title(f"Contagem de valores — {coluna}")
    plt.xlabel(coluna)
    plt.ylabel("Frequência")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

def plot_matriz_correlacao(df: pd.DataFrame) -> None:
    numericas = df.select_dtypes(include=np.number)

    if numericas.shape[1] < 2:
        print("Não há pelo menos duas variáveis numéricas para calcular correlação.")
        return

    corr = numericas.corr()

    plt.figure(figsize=(8, 6))
    plt.imshow(corr, aspect="auto")
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.title("Matriz de correlação — variáveis numéricas")
    plt.tight_layout()
    plt.show()

# Exemplos de uso:

plot_distribuicao_numerica(df, "Age")
plot_distribuicao_numerica(df, "Sleep_Hours_Per_Night")
plot_distribuicao_numerica(df, "Mental_Health_Score")
plot_distribuicao_numerica(df, "Avg_Daily_Usage_Hours")
plot_contagem_categorica(df, "Gender")
plot_contagem_categorica(df, "Academic_Level")
plot_contagem_categorica(df, "Country")
plot_contagem_categorica(df, "Most_Used_Platform")
plot_contagem_categorica(df, "Affects_Academic_Performance")
plot_contagem_categorica(df, "Overall_Impact")
dados_sem_id= df.drop(columns='Student_ID')
plot_matriz_correlacao(dados_sem_id)
```

```python
# Detecção simples de outliers em variáveis numéricas usando IQR
# Esta função é um exemplo. O grupo deve avaliar se o critério faz sentido para cada atributo.

def detectar_outliers_iqr(df: pd.DataFrame, coluna: str) -> pd.DataFrame:
    serie = df[coluna].dropna()

    q1 = serie.quantile(0.25)
    q3 = serie.quantile(0.75)
    iqr = q3 - q1

    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr

    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]

    print(f"Coluna: {coluna}")
    print(f"Q1: {q1}")
    print(f"Q3: {q3}")
    print(f"IQR: {iqr}")
    print(f"Limite inferior: {limite_inferior}")
    print(f"Limite superior: {limite_superior}")
    print(f"Número de outliers: {len(outliers)}")

    return outliers

# Exemplo de uso:

outliers = detectar_outliers_iqr(df, "Age")

outliers = detectar_outliers_iqr(df, "Sleep_Hours_Per_Night")

outliers = detectar_outliers_iqr(df, "Mental_Health_Score")

outliers = detectar_outliers_iqr(df, "Avg_Daily_Usage_Hours")
```

Além de não contribuir para os modelos, sua permanência poderia introduzir ruído nos algoritmos de clustering e aumentar riscos relacionados à identificação indireta dos indivíduos.

Não foi necessário realizar tratamento de valores ausentes, pois a análise inicial indicou ausência de dados nulos no dataset. Isso reduz a necessidade de técnicas de imputação e evita possíveis distorções estatísticas decorrentes da substituição artificial de valores.

As variáveis categóricas, como `Gender`, `Academic_Level`, `Country`, `Most_Used_Platform` e `Overall_Impact`, serão transformadas utilizando técnicas de codificação, principalmente One-Hot Encoding. Essa etapa é necessária porque algoritmos como K-Means e regressão linear trabalham adequadamente apenas com dados numéricos. A codificação permite preservar a informação sem impor relações ordinais artificiais entre categorias.

As variáveis numéricas serão normalizadas utilizando técnicas como Z-score Standardization ou Min-Max Scaling. A normalização é importante porque atributos como horas de uso, sono e indicadores de saúde mental possuem escalas diferentes. Sem esse processo, variáveis com valores maiores poderiam exercer influência desproporcional nos cálculos de distância utilizados pelo clustering.

Não foi realizada remoção inicial de outliers, pois as análises exploratórias não indicaram presença significativa de valores extremos capazes de comprometer os modelos. A decisão de manter esses registros busca preservar possíveis variações comportamentais relevantes para interpretação dos perfis encontrados. Ainda assim, a influência de valores extremos será monitorada durante a modelagem estatística.

Algumas dessas decisões introduzem riscos e limitações. A aplicação de One-Hot Encoding pode aumentar a dimensionalidade dos dados, especialmente em variáveis com muitas categorias, o que pode dificultar a interpretação dos clusters. Além disso, a normalização pode reduzir diferenças naturais entre atributos caso aplicada de maneira inadequada. Já a manutenção de categorias raras pode gerar grupos pouco representativos estatisticamente.

As decisões de preparação também estão diretamente relacionadas às diretrizes de IA responsável escolhidas no projeto. A remoção de `Student_ID` contribui para a diretriz de Privacidade e Segurança, reduzindo riscos de identificação dos participantes. Da mesma forma, o cuidado no tratamento de variáveis relacionadas à saúde mental e hábitos digitais busca evitar interpretações inadequadas sobre indivíduos específicos.

Em relação à Explicabilidade, a escolha por técnicas de preparação simples e interpretáveis, como normalização padronizada e One-Hot Encoding, favorece a compreensão dos modelos utilizados e facilita a interpretação dos resultados obtidos nos clusters e regressões.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

def preparar_dados(
    df: pd.DataFrame,
    colunas_remover=None,
    aplicar_normalizacao=True
):

    df_prep = df.copy()

    # 1. Remover duplicatas
    df_prep = df_prep.drop_duplicates()

    # 2. Processar a coluna "Country"

    # Descobrir quais são os 5 países com mais estudantes no dataset
    top_10_paises = df_prep['Country'].value_counts().nlargest(10).index

    # Manter o nome se for top 5. Se não for, renomear para 'Other'
    df_prep['Country'] = df_prep['Country'].apply(lambda x: x if x in top_10_paises else 'Other')

    categorias_country = [
    'Other',
    'Australia',
    'Canada',
    'France',
    'Germany',
    'India',
    'Mexico',
    'Spain',
    'UK',
    'USA'
    ]

    df_prep['Country'] = pd.Categorical(
        df_prep['Country'],
        categories=categorias_country,
        ordered=False
    )

    # 3. Remover colunas irrelevantes 
    if colunas_remover is None:
        colunas_remover = ["Student_ID"]

    colunas_existentes = [c for c in colunas_remover if c in df_prep.columns]
    df_prep = df_prep.drop(columns=colunas_existentes)

    # 4. Guardar as colunas numéricas originais
    colunas_numericas_originais = df_prep.select_dtypes(
        include=["int64", "float64"]
    ).columns.tolist()

    # 5. Identificar colunas categóricas
    colunas_categoricas = df_prep.select_dtypes(
        include=["object", "category"]
    ).columns

    # 6. Separar categóricas binárias e multiclasse
    colunas_binarias = [
        col for col in colunas_categoricas
        if df_prep[col].nunique(dropna=False) == 2
    ]

    colunas_multiclasse = [
        col for col in colunas_categoricas
        if df_prep[col].nunique(dropna=False) > 2
    ]

    # 7. Label Encoding para binárias
    encoders = {}

    for col in colunas_binarias:
        le = LabelEncoder()
        df_prep[col] = le.fit_transform(df_prep[col].astype(str))

        encoders[col] = dict(
            zip(le.classes_, le.transform(le.classes_))
        )

    # 8. One-Hot Encoding para multiclasse
    if colunas_multiclasse:
        df_prep = pd.get_dummies(
            df_prep,
            columns=colunas_multiclasse,
            drop_first=True,
            dtype=int
        )

    # 9. Normalizar apenas as numéricas originais
    if aplicar_normalizacao and colunas_numericas_originais:
        scaler = StandardScaler()

        df_prep[colunas_numericas_originais] = scaler.fit_transform(
            df_prep[colunas_numericas_originais]
        )

    return df_prep, encoders


# Aplicação
df_prep, encoders = preparar_dados(df)

display(df_prep.head())

print("Mapeamentos das variáveis binárias:")
for col, mapa in encoders.items():
    print(f"{col}: {mapa}")
```
